### 4.3 Работа с датасетом "house price"

##### 4.1 Исследуйте, какие характеристики недвижимости влияют на стоимость недвижимости, с применением не менее 5 диаграмм из урока.
- Анализ сделайте в формате storytelling: дополнить каждый график письменными выводами и наблюдениями.

1. Для лучшего анализа зависимости цены домов от параметров домов лучше всего построить матрицу корреляции.

```python
cor = df.corr(numeric_only=True)
cor = np.round(cor, 2)
cor = np.abs(cor)
```

```python
plt.figure(figsize=(15, 10))

sns.set(font_scale=1.1)
sns.heatmap(cor, annot=True, linewidths=0.5, cmap='coolwarm')
```

Из данной матрицы видно, что цена больше всего зависит от жилой площади, от оценки дома и от площади надземных сооружений

2. Провдём анализ зависимости цены от жилой площади

```python
plt.figure(figsize=(15, 10))
plt.bar(df['sqft_living'], df['price'], ec="red")
plt.xlabel('Жилая площадь')
plt.ylabel('Стоимость')
plt.title('Зависимость стоимости от жилой площади')
```

Из гафика видна прямая зависимость: чем больше жилая площадь, тем выше стоимость домов

3. Проверим зависемость цены от площади надземных сооружений

```python
sns.jointplot(x=df['sqft_above'], y=df['price'], kind='reg');
```

Здесь так же прослеживается прямая зависимость.

4. Проверим третью зависимость: цена от оценки дома

```python
plt.figure()
sns.boxplot(x=df['grade'], y=df['price'], whis=3)
plt.xlabel('Оценка')
plt.ylabel('Стоимость')
plt.title('Зависимость стоимости от оценки дома');
```

Здесь также видно, что оценка дома сильно влияен на цену.

5. Для проверки матрицы корреляции проверим слабую зависимость цены от состояния домов.

```python
sns.jointplot(x=df['condition'], y=df['price'], kind='reg');
```

Из графика видно, что состояние домов практически не влияет на их цену
